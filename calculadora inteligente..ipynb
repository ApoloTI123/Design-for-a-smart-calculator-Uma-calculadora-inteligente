```python
historico = []

while True:
  print ("\n=== Calculadora Inteligente ===")
  print ("1- soma")
  print ("2- subtração")
  print ("3- multiplicação")
  print ("4- divisão")
  print ("5- ver histórico")
  print ("0- sair")

  opção = input("Escolha uma operação: ")

  if opção == "0":
    print ("encerrando calculadora...")
    break

  elif opção == "5":
    if historico:
      print ("\n=== histórico ===")
      for item in historico:
        print(item)
    else:
        print("nenhuma operação realizada ainda.")

  elif opção in ["1", "2", "3", "4"]:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))

    if opção == "1":
      resultado = num1 + num2
      operacao_texto = f"{num1} + {num2} = {resultado}"

    elif opção == "2":
      resultado = num1 - num2
      operacao_texto = f"{num1} - {num2} = {resultado}"

    elif opção == "3":
      resultado = num1 * num2
      operacao_texto = f"{num1} * {num2} = {resultado}"

    elif opção == "4":
      if num2 != 0:
        resultado = num1 / num2
        operacao_texto = f"{num1} / {num2} = {resultado}"

      else:
        print("erro: divisão por zero!")
        continue

    print("Resultado:", resultado)
    historico.append(operacao_texto)

else:
  print("opção inválida! tente novamente.")




```
